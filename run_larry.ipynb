```python
import pandas as pd
import numpy as np
import quantstats as qs

def make_index_fixed(ticker_weights, rebalance="1M", period="max", returns=None, match_dates=False):
    index = None
    portfolio = {}
    
    #   
    for ticker in ticker_weights.keys():
        if (returns is None) or (ticker not in returns.columns):
            ticker_returns = qs.utils.download_returns(ticker, period)  #   .
        else:
            ticker_returns = returns[ticker]  #   ,   .
        
        portfolio[ticker] = ticker_returns  #   .
    
    index = pd.DataFrame(portfolio).dropna()  #   .
    
    #   .
    for col in index.columns:
        if index[col].dtype == 'object':
            index[col] = pd.to_numeric(index[col], errors='coerce')
    
    if match_dates:
        index = index[max(index.ne(0).idxmax()):]
    
    if rebalance is None:
        for ticker, weight in ticker_weights.items():
            index[ticker] = weight * index[ticker]
        return index.sum(axis=1)
    
    last_day = index.index[-1]
    rbdf = index.resample(rebalance).first()
    rbdf["break"] = rbdf.index.strftime("%s")
    index = pd.concat([index, rbdf["break"]], axis=1)
    index["first_day"] = pd.isna(index["break"]) & ~pd.isna(index["break"].shift(1))
    index.loc[index.index[0], "first_day"] = True
    
    for ticker, weight in ticker_weights.items():
        index[ticker] = np.where(index["first_day"], weight * index[ticker], index[ticker])
    
    # 'break'  .
    if 'break' in index.columns:
        index.drop(columns=['break'], inplace=True)
    
    index.drop(columns=["first_day"], inplace=True)
    index.dropna(how="all", inplace=True)
    
    return index[index.index <= last_day].sum(axis=1)

#    fmaga DataFrame .
# tickers period    .
tickers = {'AAPL': 0.5, 'AMZN': 0.5}  # : Apple Amazon  50%  
fmaga = make_index_fixed(tickers, period='2y')

#  .
print(fmaga.head())
```

```python
qs.reports.html(fmaga, "^KS11", title='GLD vs KOSPI3200', output='output/ticker_vs_KOSPI200.html')
```
